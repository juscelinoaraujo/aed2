# Análise de Redes de Coautoria na UFRN

## Importação de bibliotecas

```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
import seaborn as sns
import requests
import io
```

## Rede envolvendo o ODS 6

### Geração do dataframe

```python
urlData = requests.get("https://raw.githubusercontent.com/juscelinoaraujo/aed2/main/U2T1/UFRN_ODS_6.csv").content
df6 = pd.read_csv(io.StringIO(urlData.decode('utf-8')))
df6.drop(columns=['Author full names', 'Title', 'Year',
       'Source title', 'Volume', 'Issue', 'Art. No.', 'Page start', 'Page end',
       'Page count', 'Cited by', 'DOI', 'Link', 'Document Type',
       'Publication Stage', 'Open Access', 'Source', 'EID'], inplace=True)
df6 = df6.rename(columns={'Authors': 'names', 'Author(s) ID': 'id'})
names_list = []
id_list = []

for paper in range(len(df6.index)):
  df6.names[paper] = df6.names[paper].split(';')
  names_list += df6.names[paper]
  df6.id[paper] = df6.id[paper].split(';')
  id_list += df6.id[paper]
```

### Construção do grafo

```python
g6 = nx.Graph()
g6.add_nodes_from(id_list)

for paper in range(len(df6.index)):
  for a1 in df6.id[paper]:
    for a2 in df6.id[paper]:
      if a1 != a2:
        g6.add_edge(a1, a2)
```

### Propriedades do grafo

```python
print(g6.number_of_nodes())                               # number of nodes
print(g6.number_of_edges())                               # number of edges
print(nx.degree_assortativity_coefficient(g6))            # degree assortativity
highest_degree_node = max(g6.nodes(), key=lambda node: g6.degree(node))
print(g6.degree(highest_degree_node))                     # highest degree
print(nx.average_clustering(g6))                          # average clustering
nx.number_connected_components(g6)                        # connected components
```

### Análise visual

```python
nx.draw(g6)
```

```python
# The code in this cell is adapted from the following notebook:https://github.com/ivanovitchm/datastructure/blob/main/lessons/week_06/assortativity.ipynb:

# average degree of neighbors
degree, avg_neigh_degree = zip(*nx.average_degree_connectivity(g6).items())

# convert to list
degree = list(degree)
avg_neigh_degree = list(avg_neigh_degree)

plt.style.use("fivethirtyeight")
fig, ax = plt.subplots(1,1,figsize=(12,8))

sns.regplot(x=degree,y=avg_neigh_degree,ax=ax)

ax.set_xlabel("Node Degree")
ax.set_ylabel("Average neigbhor degree")
ax.set_xlim(0,65)

# Save figure
plt.savefig("degree_assortativity.png",
            format="png",
            dpi=400,
            bbox_inches="tight",
            transparent=True)

plt.show()
```
